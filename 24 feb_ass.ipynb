```python
Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.
For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.
```

```python
import pandas as pd
def calculate_sum_of_first_three_values(df):
    values_column = df['Values']
    sum_of_first_three = sum(values_column[:3])
    print("Sum of the first three values:", sum_of_first_three)
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
calculate_sum_of_first_three_values(df)
```

```python
Q5. How are DataFrame.size() and DataFrame.shape() different?
```

```python
DataFrame.size:

The size attribute of a DataFrame returns the total number of elements in the DataFrame.
It represents the product of the number of rows and the number of columns in the DataFrame.
The size attribute does not provide information about the structure or dimensions of the DataFrame.
The size attribute returns a single integer value representing the total number of elements in the DataFrame.

DataFrame.shape:

The shape attribute of a DataFrame returns a tuple that represents the dimensions of the DataFrame.
It provides the number of rows and columns in the DataFrame in the format (num_rows, num_columns).
The shape attribute is commonly used to check the size or structure of the DataFrame.
The shape attribute returns a tuple with two values, representing the number of rows and columns in the DataFrame, respectively.
```

```python
import pandas as pd
df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df.size
```

```python
import pandas as pd
df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df.shape
```

```python
import pandas as pd
df = pd.DataFrame({'A':[3,8,6,2,9],
                   'B':[5,2,9,3,1],
                   'C':[1,7,4,5,2]})
selected_rows =  df[(df['A'] > 5) & (df['B'] < 10)]
selected_rows
```

```python
import pandas as pd
df = pd.DataFrame({'A':[3,8,6,2,9],
                   'B':[5,2,9,3,1],
                   'C':[1,7,4,5,2]})
def select_rows(df):
    
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]

    
    return selected_rows

new_df = select_rows(df)
new_df
```

```python
Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.
```

```python
import pandas as pd
date  = pd.date_range(start = '2023-04-25',end = '2023-05-05')
df_date = pd.DataFrame({'date':date})
df_date['Sales']= [400,300,500,450,561,600,800,400,200,120,320]
df_date['MovingAverage'] = df_date['Sales'].rolling(window=7, min_periods=1).mean()
df_date
```
